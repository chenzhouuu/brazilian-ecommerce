```python
import pandas as pd
import numpy as np
```

```python
# import geolocation data
olist_geolocation_data = pd.read_csv('olist_geolocation_dataset.csv',\
                                    dtype={'geolocation_zip_code_prefix':str})
print(len(olist_geolocation_data))
```

```python
olist_geolocation_data.head()
```

```python
olist_geolocation_data['geolocation_zip_code_prefix_1_digits'] = olist_geolocation_data['geolocation_zip_code_prefix'].str[0:1]
olist_geolocation_data['geolocation_zip_code_prefix_2_digits'] = olist_geolocation_data['geolocation_zip_code_prefix'].str[0:2]
olist_geolocation_data['geolocation_zip_code_prefix_3_digits'] = olist_geolocation_data['geolocation_zip_code_prefix'].str[0:3]
olist_geolocation_data['geolocation_zip_code_prefix_4_digits'] = olist_geolocation_data['geolocation_zip_code_prefix'].str[0:4]
```

```python
olist_geolocation_data.head()
```

```python
olist_geolocation_data['geolocation_zip_code_prefix'].value_counts().\
to_frame().describe()
```

```python
# Removing some outliers
#Brazils most Northern spot is at 5 deg 16′ 27.8″ N latitude.;
olist_geolocation_data = olist_geolocation_data[olist_geolocation_data.geolocation_lat <= 5.27438888]
#it’s most Western spot is at 73 deg, 58′ 58.19″W Long.
olist_geolocation_data = olist_geolocation_data[olist_geolocation_data.geolocation_lng >= -73.98283055]
#It’s most southern spot is at 33 deg, 45′ 04.21″ S Latitude.
olist_geolocation_data = olist_geolocation_data[olist_geolocation_data.geolocation_lat >= -33.75116944]
#It’s most Eastern spot is 34 deg, 47′ 35.33″ W Long.
olist_geolocation_data = olist_geolocation_data[olist_geolocation_data.geolocation_lng <=  -34.79314722]
```

```python
# treat the latitute and longitude coordinates and transform then to Mercator x/y Coordinates.
from datashader.utils import lnglat_to_meters as webm
```

```python
X,y = webm(olist_geolocation_data.geolocation_lat,olist_geolocation_data.geolocation_lng)
print(X,y)
```

```python
olist_geolocation_data['X'] = pd.Series(X)
olist_geolocation_data['y'] = pd.Series(y)
```

```python
olist_geolocation_data.head()
```

```python
olist_geolocation_data['geolocation_zip_code_prefix'] = olist_geolocation_data['geolocation_zip_code_prefix'].astype(int)
olist_geolocation_data['geolocation_zip_code_prefix_1_digits'] = olist_geolocation_data['geolocation_zip_code_prefix_1_digits'].astype(int)
olist_geolocation_data['geolocation_zip_code_prefix_2_digits'] = olist_geolocation_data['geolocation_zip_code_prefix_2_digits'].astype(int)
olist_geolocation_data['geolocation_zip_code_prefix_3_digits'] = olist_geolocation_data['geolocation_zip_code_prefix_3_digits'].astype(int)
olist_geolocation_data['geolocation_zip_code_prefix_4_digits'] = olist_geolocation_data['geolocation_zip_code_prefix_4_digits'].astype(int)
```
